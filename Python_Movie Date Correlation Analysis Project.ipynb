```python
# Import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Read in the data
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Check to see if there's missing value

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Drop the missing value, which is only a small proportion of the dataframe
df = df.dropna()
```

```python
# Double check if there's missing value
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Confirm the data type of our col
df.dtypes
```

```python
#Change the data type of our col to make it more readibale
df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df.head()
```

```python
#create a correct Year col
#() -> indicate a capture group, meaning the content matched by this group will be extracted.
#[0-9] -> Matches any digit from 0 to 9.
#{4} -> Matches the previous pattern exactly 4 times, which means that it matches a sequence of 4 digits.
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
# Add this code to allow us to scroll and see all rows
# pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# drop any duplicates values
df.drop_duplicates()

# if you only want to drop the duplicate value in certain col, use the following code
# df['company'] = df['company'].drop_duplicates()
```

```python
# Hypothesis: Budget would probably have high correlation to the gross 
# Use scatter plot to see the correlation between budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross using seaborn
# having a line of best fit to help expresses the relationship between the points

sns.regplot(x='budget',y='gross', data=df, scatter_kws={'color':'red'} , line_kws={'color':'green'} )
```

```python
# Pearson correlation is used for continuous data that follows a normal distribution & assumes a linear relationship between variables
# Kendall and Spearman correlations are used for non-parametric data that may not follow a normal distribution and do not assume a linear relationship between variables 
df.corr(method='pearson') 
```

```python
# Prove the hypothesis of strong correlation between budget and gross
# Use the heatmap to show the high correlation between budget and gross
correlation_matrix = df.corr(method='pearson') 
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# look at the correlation between votes and gross
# unstack shows all the columns compare to each one

correlation_mat = df.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Sort out the pairs, and only show the high correlation ones
sorted_pairs = corr_pairs.sort_values(ascending=False)
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

### Conclusion: votes & budget have the highest correlation to gross earnings
